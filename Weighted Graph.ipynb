```python
import requests #library to fetch the 
import numpy as np
import networkx as nx #library for graph creation
import pickle #library to save the graph
from itertools import permutations # to create the edges between nodes
```

## Extract Tags
##### This class reads the json file from the url. And saves extracted tags in a list. 

```python

class extract_tags:
    def __init__(self, amt):
        self._amt = amt
        self.__url = None
        self.__json = None
        self.unique_tags = set()
        self.tags = []
    def read_json(self):
        for i in range(1, self._amt + 1):
            self.__url = 'https://api.stackexchange.com/2.2/questions?page=' + str(i) + '&pagesize=100&order=desc&sort=activity&site=stackoverflow' 
            self.__json = requests.get(self.__url).json()['items']
            for it in range(len(self.__json)):
                self.tags.append(self.__json[it]['tags'])
                for x in self.__json[it]['tags']:
                    self.unique_tags.add(x) 
```

## Create_graph class
#### This is the child class of the above class. Is uses the list in which tags are stored. And Further create the graph based on that list. 

```python
class create_graph(extract_tags):
    def __init__(self, amt):
        self.graph = nx.Graph()
        self.amt = amt
        super().__init__(self.amt)
        self.read_json()
        self.graph.add_nodes_from(self.unique_tags)
    def add_weight(self):
        for tag in self.tags:
            self.__perm = list(permutations(tag, 2))
            for all_perm in self.__perm:
                if all_perm in self.graph.edges:
                    self.graph.edges[all_perm]['weight'] += 1
                else:
                    self.graph.add_weighted_edges_from([(all_perm[0], all_perm[1], 1)])
```

#### We created a object of create_graph class and passed the number of pages.

```python
cls = create_graph(100)
```

#### Add weight function of the create_graph class will add weights to the graph

```python
cls.add_weight()
```

###### This is the command that I used to order the recommanded tags.

```python
list(dict(sorted(dict(cls.graph['python']).items(), key = lambda x: x[1]['weight'], reverse = True)).keys())
```

```python
pickle.dump(cls.graph, open('graph.pkl', 'wb'))
```
